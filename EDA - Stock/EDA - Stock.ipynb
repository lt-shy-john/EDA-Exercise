# EDA of Stock Price Dataset

Stock price provides wealth of data and there is a mature industry to do so. 

This EDA is conducted using __Python__. In the following let us import the libraries. 

```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

We import the data file as below. 

```python
price_df = pd.read_csv('prices.csv', header=None)
print(f'{price_df.shape[0]} rows has been imported. ')
```

The data are stored in data frames with rows as the stock and columns are each closing price. The price starts from 1st Jan 2014 to Feb 2021. 

```python
price_df.head(10)
```

We also look into the descriptive statistics of the data set, as below. 

```python
price_df.transpose().describe()
```

```python
print('=== Coefficient of variation ===')
for i in range(20): 
    print('{}: {}'.format(i, price_df.iloc[i,:].std(), price_df.iloc[i,:].mean()))
```

We can see the following: 
* There are no missing values, as from the number of rows is the same as the number in all columns. 
* The mean value indicates where the stock is at around these 6 years. Different stocks have different mean values. 
* The coefficient of variation differs a lot between stocks, but all has value greater than 1 means they are quite high in risk. 
* The median often not deviates with the mean value too much ($\pm10$). 

Just in case the data contains missing values, we shall fill them with $0. 

```python
# Fill missing values as 0 
price_df.fillna(value=0, inplace=True)
# price_df
```

As a start we visualise the time series of all stocks in this dataset. 

```python
plt.figure(figsize=(20,10))
plt.title('Stock Price since 1st Jan 2014')
plt.ylabel('USD')
plt.xlabel('Days')
plt.tight_layout()
for i in range(len(price_df)): 
    plt.plot(price_df.iloc[i])
plt.show()
```

From each stocks, we can see that
* Most of them have several peaks. 
* They does not seemed symmetric, even by looking at individual peaks. 

```python
plt.figure(figsize=(20,10))
plt.suptitle('Distribution of Stock Price since 1st Jan 2014')
plt.ylabel('USD')
plt.xlabel('Days')
plt.tight_layout()
for i in range(len(price_df)): 
    plt.subplot(5, 4, i+1)
    sns.histplot(price_df.iloc[i], kde=True)
plt.show()
```

We also wanted to see outliers, which can be shown from boxplots. 

```python
plt.figure(figsize=(20,10))
plt.suptitle('Distribution of Stock Price since 1st Jan 2014')
plt.tight_layout()
for i in range(len(price_df)): 
    plt.subplot(4, 5, i+1)
    plt.boxplot(price_df.iloc[i], labels=[price_df.index[i]])
plt.show()
```

As we can see, stocks 2, 11, 12 and 14 has quite substantial numbers of outliers. More on the higher price ranges. 

We can also see some traces of multicollinearity of stocks 10, 13, 14, 17, 2, 19, 1, 0 and 9. Clusters of mid to high correlation clustered between stocks 3, 5, 12, 18, 4, 6 and 11. This means if we trying to do a regression model, we may select one from each cluster as a factor. 

```python
sns.clustermap(price_df.transpose().corr(),figsize=(20,20),annot=True).fig.suptitle('Correlation Matrix')
plt.show()
```
